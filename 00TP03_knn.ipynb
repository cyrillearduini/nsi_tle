### Trouver la maison majoritaire

- La fonction `frequence_des_maisons` suivante renvoie un dictionnaire contenant la liste des maisons et le nombre d'élèves de chacune :

````python
# A exécuter
def frequence_des_maisons(table):
    freq={}
    for eleve in table:
        maison=eleve["maison"]
        if maison in freq.keys():
            freq[maison] +=1
        else:
            freq[maison] =1
    return freq
````

### Trouver les $k$ plus proches voisins

Pour déterminer la maison de l'élève, nous décidons d'utiliser les $7$ plus proches voisins.

- La fonction `voisins` prend en paramètre un nouvel élève et une table et renvoie la table passée en paramètre triée par distance croissante. Pour cela, on utilise la fonction sorted sur une table en précisant le critère calculé par la fonction `dist_nouv`.


````python
# A exécuter
def voisins(nouveau,table):
    # On crée une fonction qui calcule la distance
    def dist_nouv(el):
        return distance(el,nouveau)
        
    # On trie la liste selon la distance
    return sorted(table,key=dist_nouv)
````

````python
# Vérification
voisins(hermione,poudlard)
````
