## Le module Turtle
Pour dessiner des formes géométriques, Python propose le module **Turtle**.  
Ce module permet de faire simplement des dessins en indiquant au curseur où et comment se déplacer.  
[Documentation Turtle python.org](https://docs.python.org/fr/3/library/turtle.html)  
Le code fourni ci dessous vous donnera les bases pour utiliser le module :  

````python
# A executer et analyser

import turtle as tl  # Import du module
tl.speed(3)          # Vitesse de dessin : 6:normal, 10:rapide, 0:instantané
# le code suivant permet de commencer le dessin bien à gauche
tl.penup()           # Lever le stylo
tl.goto(-300, 120)   # Aller à la position 
tl.pendown()         # Baisser le stylo

tl.forward(200)      # Avancer de 200 pixels
tl.right(90)         # Tourner de 90° à droite
tl.forward(100)      # Avancer de 100 pixels
tl.left(45)          # Tourner de 45° à gauche
tl.forward(100)      # Avancer de 100 pixels
tl.left(135)         # Tourner de 135° à gauche
tl.forward(200)      # Avancer de 200 pixels

tl.done()            # Lancer le dessin (spécifique à Basthon)
````
