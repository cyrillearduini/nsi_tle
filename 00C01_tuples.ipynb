En classe de première, nous avons vu les différents types simples permettant de stocker de l’information : **int**, **float**, **bool**.  
Le type **str** est aussi un type simple mais on peut sur ce type particulier accéder aux éléments de celui-ci. Par exemple si on crée une chaîne avec la commande ```ch='bonjour'``` alors l’expression ```ch[0]``` nous renverra la valeur ```'b'``` et l’expression ```ch[3:7]``` la valeur ```'jour'``` valeurs qui seront également du type str.  

Avec ces types simples, nous pouvons définir des variables qui représenteront chacune **une** donnée avec **une** valeur. Ceci est suffisant pour traiter un petit nombre de données, mais lorsque nous allons vouloir traiter un plus gros volume de données, ceci peut devenir contraignant complexe à gérer.  
Par exemple, si nous voulons gérer un ensemble de points en géométrie, chacun comporte une abscisse et une ordonnée, il nous faudra donc deux valeurs par objet, si nous gérons un train, il sera composé d’une locomotive et d’un ou plusieurs wagons (dont nous ne connaissons pas le nombre à l’avance).  

Pour répondre à ce type de besoins, nous allons donc étudier trois nouveaux types d’objets : le type **tuple** (n-uplet), le type **list** (liste) et le type **dict** (dictionnaire). Ces types présentent des **points communs** mais aussi des **différences**, ils nous permettront d’adresser des besoins similaires et leur différences nous aideront à faire notre choix en fonction du problème à résoudre.


---
## Définition

Un objet de type tuple est une suite ordonnée d’éléments qui peuvent être chacun de n’importe quel type. C’est une séquence d’éléments non modifiables.

---
## Création

Pour créer un tuple, on utilise les parenthèses et on sépare les éléments par des virgules :

````python
t = (3,7)
````

````python
t
````

````python
type(t)
````

Un tuple peut être vide

````python
t=()
````

````python
t
````

Ou contenir une seule valeur

````python
t=(4,)
````

````python
t
````

````python
type(t)
````

Attention à ne pas oublier la virgule, sinon le type sera différent

````python
t=(4)
````

````python
t
````

````python
type(t)
````

Il peut également contenir des objets de types différents

````python
t = (2,3,'a',8,)
````

````python
t
````

Un tuple peut contenir un autre tuple

````python
t = (2,3,'a',(6,7))
````

````python
t
````

La fonction ```tuple``` permet à partir d’une chaîne de caractères ou d’une liste (que nous verrons plus tard) de créer un tuple :

````python
t = tuple('Première NSI')
````

````python
t
````

## Accès aux éléments

Pour accéder aux éléments d’un tuple, on utilise les crochets :

````python
t = (2,5,6,3,0,3,6)
````

````python
t[2]
````

- Comme pour les chaines de caractères, les indices des éléments commencent à 0 et se terminent à n-1 si n est le nombre d’éléments du tuple.  

  
- On accèdera à la longueur d’un tuple avec la fonction ```len```

````python
len(t)
````

- Un indice **négatif** permet de parcourir le tuple en partant de la fin (en commençant à -1)

````python
t[-2]
````

- On peut accéder à un **tranche** de tuple en utilisant la syntaxe suivante :

````python
t[2:5]
````

````python
t[:5]
````

````python
t[3:]
````

````python
t[:3] + (10,) + t[4:]
````

- Pour un tuple contenant un autre tuple, il faudra doubler les indices :

````python
t = ( ('a','b') , ('c','d') )
````

````python
t
````

````python
t[0]
````

````python
t[1][0]
````

- Pour parcourir les éléments, on peut utiliser une boucle ```for```, on dit que les tuples sont **itérables** :

````python
t = (3,5,12)
for elt in t:
    print (elt, end=' ')
````

- Le tuple peut être également parcouru par ses indices :

````python
for i in range(len(t)):
    print (t[i], end=' ')
````

- Pour tester l’appartenance d’un élément à un tuple, on utilise l’opérateur ```in``` :

````python
t = ('a','b','c')
````

````python
'a' in t
````

````python
'd' in t
````

- On ne peut pas modifier les éléments d’un tuple par affectation, on dit que les tuples sont **non mutables** :

````python
t[2] = 3
````

## Affectations multiples

La commande ```a,b = 1,2``` affecte 1 à la variable ```a``` et 2 à la variable ```b```. Ceci est en fait lié au fait que Python va affecter au tuple ```(a,b)``` la valeur du tuple ```(1,2)```.
Cette propriété est très utile pour permuter les valeurs de deux variables sans passer par une variable intermédiaire avec la commande ```a,b = b,a```

````python
a,b = 1,2

a,b = b,a
````

````python
a
````

````python
b
````
