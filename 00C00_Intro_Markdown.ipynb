## Le Markdown


### C'est quoi ? 
Le markdown est un langage de balisage léger, il permet de formater du texte par balisage (un peu comme le HTML)

### Pourquoi l'utiliser
- Il est plus simple pour mettre en forme du texte que *Word* ou un traitement de texte
- On peut l'utiliser **en ligne** et **gratuitement**

Dans une cellule markdown, on peut mettre écrire des ```lignes de code```
mais on ne pourra pas les ~~exécuter~~

On peut ajouter des liens hypertexte : [Google](http://google.com).
