Un objet de type list est une suite ordonnée d’éléments qui peuvent être chacun de n’importe quel type. C’est une séquence d’éléments modifiables.

## Accès aux éléments

Pour accéder aux éléments d’une liste, on utilise les crochets :

````python
l = [2,5,6,3,0,3,6]
````

````python
l[2]
````

- Comme pour les chaines de caractères, les indices des éléments commencent à 0 et se terminent à n-1 si n est le nombre d’éléments de la liste.  

  
- On accèdera à la longueur d’une liste avec la fonction ```len```

````python
len(l)
````

- Un indice **négatif** permet de parcourir la liste en partant de la fin (en commençant à -1)

````python
l[-2]
````

- On peut accéder à un **tranche** de liste en utilisant la syntaxe suivante :

````python
l[2:5]
````

````python
l[:5]
````

````python
l[3:]
````

````python
l[:3] + [10] + l[4:]
````

- Pour une liste contenant une autre liste, il faudra doubler les indices :

````python
l = [ ['a','b'] , ['c','d'] ]
````

````python
l
````

````python
l[0]
````

````python
l[1][0]
````

- Pour parcourir les éléments, on peut utiliser une boucle ```for```, on dit que les listes sont **itérables** :

````python
l = [3,5,12]
for elt in l:
    print (elt, end=' ')
````

- La liste peut être également parcouru par ses indices :

````python
for i in range(len(l)):
    print (l[i], end=' ')
````

- Pour tester l’appartenance d’un élément à une liste, on utilise l’opérateur ```in``` :

````python
l = ['a','b','c']
````

````python
'a' in l
````

````python
'd' in l
````

- Pour ajouter des éléments, on peut utiliser la méthode ```append``` :

````python
l1 = [2,5]
l1.append(2)
l1.append(3)
````

````python
l1
````

````python
l=[4, 1, 2, 6, 7, 15, 7]
del l[0]
````

````python
l
````

````python
l
````

- Pour insérer des éléments, on peut utiliser la méthode ```insert``` en lui passant en paramètres la position et l'élément à insérer :

````python
l=[4, 1, 2, 6, 7, 15, 7]
l.insert(3,99)
````

````python
l
````

Python permet également de générer une liste en appliquant une fonction à un ensemble de
valeurs.  
Cela permet de créer une liste en une seule ligne au lieu de passer par une boucle.  
Cette méthode est appelée **construction par compréhension**

````python
l=[x**2 for x in range(10)]
````

````python
l
````

Revient à écrire en une seule ligne le code suivant :

````python
l=[]
for x in range(10):
    l.append(x**2)
````

````python
l
````

Il est même possible de filtrer la liste obtenue à l’aide d’une condition !  
On peut ainsi par exemple obtenir la liste des entiers inférieurs à 50 qui sont des multiples de 3 mais pas
de 4 en une seule commande :

````python
l=[x for x in range(50) if x%3==0 and x%4!=0]
````

````python
l
````

## Exemple : tracé de courbes


Cette méthode est notammment utilisée pour tracer les représnetations graphiques de fonctions :
- On déclare une fonction correspondant à la fonction à tracer,
- On crée une liste contenant les abscisses des points a tracer,
- On crée la liste des ordonnées par compréhension,
- Il ne reste plus qu'à positionner les points.

````python
# Déclaration de la fonction à tracer
def f(x):
    return x ** 2 + x - 4

# Génération d'une liste d'ascisses (de -3 à 3 avec un pas de 0,1)
liste_x = [0.1 * n for n in range(-30, 31)] 

# Génération d'une liste d'ordonnées par compréhension (y = f(x))
liste_y = [f(x) for x in liste_x]  
````

````python
liste_x
````

````python
liste_y
````

````python
# Import du module pour tracer 
import matplotlib.pyplot as plt

# Tracé de la courbe
plt.plot(liste_x, liste_y)
plt.show()
````
