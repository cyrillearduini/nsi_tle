### ðŸ’» EXERCICE 2
>Dans la classe `Graphe_mat` du cours reprise ci-dessous, ajoutez une mÃ©thode `voisins(self,i1,i2)` qui renverra `True` si les sommets d'indices `i1` et `i2` sont adjacents et `False` sinon.

```python
class Graphe_mat() :
    ''' Classe reprÃ©sentant un graphe avec une matrice d'adjacence '''
    def __init__(self, n):
        ''' Constructeur : n = taille de la matrice, adj = matrice nxn contenant des 0 '''
        self.n = n
        self.adj = [ [0]*n for x in range (n)]

    def ajouter_arete(self,i1,i2):
        ''' Ajoute 1 dans la matrice pour reprÃ©senter l'arÃªte s1 -> s2 '''
        self.adj[i1][i2] = 1
        
    # Ã  complÃ©ter
    
```

```python
# vÃ©rification
gv=Graphe_mat(3)       # On crÃ©e un graphe avec 3 sommets
gv.ajouter_arete(0,1)  # Une arÃªte entre 0 et 1
gv.ajouter_arete(0,2)  # Une arÃªte entre 0 et 2
                       # Pas d'arÃªte entre 1 et 2

print(gv.voisins(0,1)) # Doit renvoyer True
print(gv.voisins(0,2)) # Doit renvoyer True
print(gv.voisins(1,2)) # Doit renvoyer False
```

### ðŸ’» EXERCICE 3
>Dans la classe `Graphe_lis` du cours reprise ci-dessous, ajoutez une mÃ©thode `voisins(self,s1,s2)` qui renverra `True` si les sommets `s1` et `s2` sont adjacents et `False` sinon.

```python
class Graphe_lis() :
    ''' Classe reprÃ©sentant un graphe avec une liste d'adjacence '''
    def __init__(self):
        ''' Constructeur : dictionnaire vide '''
        self.adj = {}

    def ajouter_sommet(self,s):
        if s not in self.adj:
            self.adj[s] = []
        
    def ajouter_arete(self,s1,s2):
        ''' Ajoute les sommets s2 et S2 puis s2 dans la liste des valeurs de s1 pour reprÃ©senter l'arÃªte s1 -> s2 '''
        self.ajouter_sommet(s1)
        self.ajouter_sommet(s2)
        self.adj[s1].append(s2)
    
    # Ã  complÃ©ter
    
```

```python
# vÃ©rification

gv2=Graphe_lis()            # On crÃ©e un graphe avec 3 sommets
gv2.ajouter_sommet('A')
gv2.ajouter_sommet('B')
gv2.ajouter_sommet('C')

gv2.ajouter_arete('A','B')  # Une arÃªte entre A et B
gv2.ajouter_arete('A','C')  # Une arÃªte entre A et C
                            # Pas d'arÃªte entre B et B

print(gv2.voisins('A','B')) # Doit renvoyer True
print(gv2.voisins('A','C')) # Doit renvoyer True
print(gv2.voisins('B','C')) # Doit renvoyer False
```
