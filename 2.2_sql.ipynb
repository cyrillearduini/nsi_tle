## Le langage SQL (_Structured Query Language_)

SQL est un langage permettant dexploiter les bases de donnes relationnelles. Il permet de rechercher, dajouter, de modifier 
ou de supprimer des donnes. C'est un langage **dclaratif** cest--dire quil permet de dcrire le rsultat escompt, sans dcrire la manire de lobtenir. Les instructions SQL scrivent donc dune manire qui ressemble  celle de phrases ordinaires en anglais. Cette ressemblance voulue vise  faciliter lapprentissage et la lecture.

>  Les **requtes** sont des instructions qui permettent d'intragir avec une base de donnes.
> - Le langage SQL nest pas sensible  la casse, mais il est prconis par convention d'crire les instructions **en majuscules**
> - On peut crire les instructions sur **plusieurs lignes** avec ou sans indentation (contrairement  Python, SQL n'est pas sensible  l'indentation)
> - Chaque requte doit tre obligatoirement termine par un point-virgule `;`
> - Pour ajouter une ligne de commentaires, il faut la commencer avec deux tirets `--`
> - Pour ajouter plusieurs lignes de commentaire, il faut les encadrer avec `/*` et `*/`
> - Les requtes sont composes de mots-cls appels aussi **clauses**

## Recherche croise sur plusieurs tables : les jointures

Pour faciliter le traitement des donnes, nous les avons spares en plusieurs tables, mais pour la lisibilit la recherche avec un `SELECT` ne sera pas toujours trs facile :

```python
SELECT * FROM VOITURE ;
```

Si je vous demande "_Quelle est la marque de l'enregistrement dont le modle est `Cayenne` ?_" Vous devrez dans un premier temps retrouver l'enregistrement concern :

```python
SELECT * FROM VOITURE WHERE modele = "Cayenne" ;
```

A la lecture du rsultat, nous savons que la marque est celle de la table `marque` donc la cl primaire _idMarque_ vaut `3`. Il faudra donc utiliser une seconde requte pour trouver son nom :

```python
SELECT * FROM MARQUE WHERE idMarque = 3 ;
```

Nous pouvons maintenant conclure en rpondant "_Porsche_"...
Les deux requtes nous ayant permis de retrouver l'information peuvent tre combines par ce qu'on appelle une **jointure**  effectue  l'aide du mot cl `JOIN`.  

Affichons le modle, la marque, la couleur et le kilomtrage de notre vhicule : 

```python
SELECT
    MARQUE.nom, VOITURE.modele, VOITURE.couleur, VOITURE.km
FROM
    VOITURE
JOIN 
    MARQUE
ON
    VOITURE.idMarque = MARQUE.idMarque
WHERE
    VOITURE.modele = "Cayenne" ;
```

- Lorsque les requtes deviennent complexes, il est de rigueur d'**indenter le code**, ceci permettra une bien meilleure visibilit et vitera les erreurs
- Comme nous affichons des donnes de plusieurs tables, vous noterez que nous avons **prfix** chaque attribut par la table concerne pour viter les confusions
- L'expression `JOIN MARQUE ON VOITURE.idMarque = MARQUE.idMarque` signifie qu'en plus des donnes de notre table `voiture`, nous allons afficher des donnes de la table `marque` (ici l'attribut _nom_) et nous prcisons la correspondance entre les deux tables (cl trangre table 1 = cl primaire table 2)
- Comme vous pouvez le voir, nous avons filtr les donnes avec une clause `WHERE`  

Il est cependant possible d'afficher touts les croisements entre les deux tables en supprimant la clause `WHERE` :

```python
SELECT
    MARQUE.nom, VOITURE.modele, VOITURE.couleur, VOITURE.km
FROM
    VOITURE
JOIN
    MARQUE
ON
    VOITURE.idMarque = MARQUE.idMarque ;
```

Il est possible d'utiliser les jointures sur plus de deux tables, et nous pourrons ainsi ajouter la ville de l'agence de chaque vhicule :

```python
SELECT
    MARQUE.nom, VOITURE.modele, VOITURE.couleur, VOITURE.km, AGENCE.ville
FROM
    VOITURE
JOIN
    MARQUE
ON
    VOITURE.idMarque = MARQUE.idMarque
JOIN
    AGENCE
ON
    VOITURE.idAgence = AGENCE.idAgence ;
```

Enfin, il est possible d'utiliser des **alias** sur les tables pour faciliter la lecture. Profitons-en galement renommer les colonnes et avoir un affichage soign:

```python
SELECT
    M.nom AS Marque, V.modele AS Modle,  V.couleur AS Couleur, V.km AS Kilomtrage, A.ville AS Agence
FROM
    VOITURE AS V
JOIN
    MARQUE AS M
ON
    V.idMarque = M.idMarque
JOIN 
    AGENCE AS A
ON
    V.idAgence = A.idAgence ;
```
