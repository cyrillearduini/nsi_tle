## Rappels
### Basthon en mode SQL
> ⚠️ Lors de l'utilisation d'un notebook SQL, assurez vous de l'avoir ouvert avec le bon noyau :
> - Soit en vérifiant que dans l'url, le noyau a été précisé : `https://notebook.basthon.fr/?kernel=sql&from=https://...`
> - Soit en sélectionnant le noyau SQL dans le menu de Basthon : `Noyau / Changer de noyau > / SQL`

### Import d'une base de données dans le notebook
Si vous avez ouvert le fichier pour la première fois via Git, vous n'avez rien à faire, la base de données est chargée dans le notebook.
>⚠️ Si vous réouvrez votre fichier, vous devez importer la base dans le Notebook, pour cela :
>- Téléchargez le fichier **2.6_soccer.db** disponible sur le dépot Git à coté du TP
>- Importez le fichier dans le notebook en faisant  `Fichier... / Ouvrir` et le sélectionnant, normalement, le notebook vous informera que le fichier est utilisable depuis SQL.

### Liste des tables
- Vous pouvez voir la liste des tables disponibles avec la commande :
``` SQL
.tables
```
- Pour une visualisation rapide des attributs d'une table : 
``` SQL
SELECT * FROM Table LIMIT 1;
```
- Vous pouvez également visualiser toutes les informations de chaque table, notamment ses clés primaires et étrangères en utilisant la requête suivante :
``` SQL
SELECT * FROM sqlite_master ORDER BY tbl_name, rootpage ;
```

---
## Exercices
> Pour chacune des requêtes pensez bien à renommer les colonnes pour que l'affichage des résultats soit le plus soigné possible.  
>
> _💡 Le tableau suivant représente les différents types dans la table Event._
>
> | type     | Libellé             |
> |:--------:|:-------------------:|
> | goal     | but                 |
> | penalty  | but sur penalty     |
> | own      | but contre son camp |
> | miss     | penalty raté        |
> | yellow   | carton jaune        |
> | yellow2  | 2e carton jaune     |
> | red      | carton rouge        |

- **💻 1. A l'aide des commandes précédentes, dessinez sur une feuille le diagramme relationnel de la base**

- **💻 2. Combien d'équipes ont participé à ce championnat ?**

- **💻 3. Listez les noms des clubs ainsi que leur date de création dans l'ordre chronologique de leur création**

- **💻 4. Combien de pénaltys ont été marqués ?**

- **💻 5. Combien de pénaltys ont été sifflés ?**

- **💻 6. Combien de cartons ont été distribués ?**

- **💻 7. Combien de buts ont été marqués ?**

- **💻 8. Affichez les 2 clubs ayant obtenu le plus grand nombre de cartons rouges ainsi que le nombre de cartons associés.**

- **💻 9. Donnez le classement des 3 clubs ayant marqué le plus grand nombre de buts ainsi que le nombre de buts associés.**

- **💻 10. Donnez le classement des 10 meilleurs buteurs, vous afficherez le nom du buteur, son club (nom court) et le nombre de buts inscrits.**

- **💻 11. Quel joueur a raté le plus de pénaltys ? Donnez son nom, son club (court) et le nombre.**

- **💻 12. A quelle date le LOSC a-t-il reçu le PSG ? Affichez également le nombre de buts inscrits lors de ce match.**

- **💻 13. Affichez le nom et club (court) des buteurs du match.**
